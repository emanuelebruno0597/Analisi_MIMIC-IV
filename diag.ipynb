converto i codici secondo il significato e li raggruppo sotto categori di diagnosi (i valori a 0 vengono classificati come "unknown")

```python
def convert_icd10_to_cat(x):
    if x == "0":
        return "unknown"
    if x[0] in ["A", "B"]:
        return "infectious_parasitic"
    if x[0] == "C":
        return "neoplasms"
    if x[0] == "D":
        try:
            if int(x[1:3]) <= 49:
                return "neoplasms"
            else:
                return "blood"
        except:
            return "unknown"
    if x[0] == "E":
        return "endocrine_nutritional"
    if x[0] == "F":
        return "mental"
    if x[0] == "G":
        return "nervous_system"
    if x[0] == "H":
        try:
            if int(x[1:3]) <= 59:
                return "eye"
            else:
                return "ear"
        except:
            return "unknown"
    if x[0] == "I":
        return "circulatory"
    if x[0] == "J":
        return "respiratory"
    if x[0] == "K":
        return "digestive"
    if x[0] == "L":
        return "skin"
    if x[0] == "M":
        return "musculoskeletal"
    if x[0] == "N":
        return "genitourinary"
    if x[0] == "O":
        return "pregnancy"
    if x[0] == "P":
        return "perinatal"
    if x[0] == "Q":
        return "chromosomal"
    if x[0] == "R":
        return "abnormal"
    if x[0] in ["S", "T"]:
        return "injury_poisoning_external"
    if x[0] in ["V", "Y", "W", "X"]:
        return "morbidity_mortality_external"
    if x[0] == "Z":
        return "health_status"
    if x[0] == "U":
        return "special"  
```

effettuo il one hot encoding sulla categoria di diagnosi per ogni riga del dataset

```python
labels = df['icd_diagnosis'].unique()
```

```python
labels2 = []
for elem in labels:
    tmp = 'icd_D_'+elem
    labels2.append(tmp)
```

```python
labels2
```

```python
def reduction(df, column, prefix):
    data = pd.get_dummies(df[column], prefix=prefix)
    
    out = pd.concat([df,data], axis = "columns")                   # Si concatena il vecchio dataframe con le nuove colonne generate
    out.drop(column, axis='columns', inplace = True)               # Si droppa la vecchia colonna utilizzata
    return out 
```

```python
test = reduction(test, 'icd_diagnosis', 'icd_D')
```

```python
test
```

eseguo il raggruppamento per id di ammissione e sommo le volte che il paziente Ã¨ stato diagnosticato con una patologia che rientra in una determinata categoria, calcolo inoltre il numero seq num massimo, minimo e medio (stando alla documentazione Ã¨ il valore indicativo della prioritÃ  e della gravitÃ  della diagnosi)

```python
a = test.groupby('hadm_id')[labels2].sum()
b = test.groupby('hadm_id')['seq_num'].min()
tmp = pd.concat([a,b], axis=1)
tmp = tmp.rename(columns={'seq_num': 'seq_min'})

c = test.groupby('hadm_id')['seq_num'].max()
tmp = pd.concat([tmp,c], axis=1)
tmp = tmp.rename(columns={'seq_num': 'seq_max'})

d = test.groupby('hadm_id')['seq_num'].mean()
tmp = pd.concat([tmp,d], axis=1)
tmp = tmp.rename(columns={'seq_num': 'seq_mean'})

tmp.reset_index(level=0, inplace=True)

test2 = tmp.copy()
```
