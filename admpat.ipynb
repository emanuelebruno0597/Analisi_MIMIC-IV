calcolo l'etÃ  dei pazienti in ogni ammissione

```python
def ageCalculator(df):
    age=[]
    for i in tqdm(range(0, len(df))):
        anchor_age = df['anchor_age'][i]
        anchor_year = df['anchor_year'][i]
        
        row = df['admittime'][i]
        year = row.split('-')[0]
        admitYear = int(year)
        
        value = anchor_age + (admitYear - anchor_year)
        
        age.append(value)
    
    df['age'] = age
```

```python
ageCalculator(result)
```

per ogni ammissione verifico che lo stesso paziente abbia un'ammissione successiva e se si calcolo quanto tempo dopo Ã¨ rientrato in ospedale e lo assegno a una delle seguenti categorie:
riammesso entro 30 giorni;
riammesso entro 90 giorni;
riammesso entro 365 giorni;
riammesso oltre i 365 giorni.

```python
def addIntoDaysColumn(lists, d30, d90, d365, d365p):
    lists[0].append(d30)
    lists[1].append(d90)
    lists[2].append(d365)
    lists[3].append(d365p)

def riadmissionCalculator(df):
    days30 = []
    days90 = []
    days365 = []
    days365plus = []
    lists = [days30, days90, days365, days365plus]
    
    #if(len(df) != 0):
    #     subject_id = df['subject_id'][0]
    #else:
    #    return
    
    for i in tqdm(range(0, len(df)-1)):
        subject_id = df['subject_id'][i]
        subject_id_succ = df['subject_id'][i+1]
        
        if (subject_id == subject_id_succ):
            disch = datetime.strptime(df['dischtime'][i],   '%Y-%m-%d %H:%M:%S')
            admit = datetime.strptime(df['admittime'][i+1], '%Y-%m-%d %H:%M:%S')
            
            diff = (admit - disch).days
            
            if(diff <= 30):
                addIntoDaysColumn(lists, 1, 0, 0, 0)
                
            elif(diff <= 90):
                addIntoDaysColumn(lists, 0, 1, 0, 0)
            
            elif(diff <= 365):
                addIntoDaysColumn(lists, 0, 0, 1, 0)
            
            else:
                addIntoDaysColumn(lists, 0, 0, 0, 1)
            
        else:
            addIntoDaysColumn(lists, 0, 0, 0, 0)
            
    addIntoDaysColumn(lists, 0, 0, 0, 0)
    
    df['riad 30d'] = days30
    df['riad 90d'] = days90
    df['riad 365d'] = days365
    df['riad +365d'] = days365plus
```

```python
riadmissionCalculator(result)
```
